```python
# Importing Matplotlib library to draw the 2 to the upper i values as a graph

import matplotlib.pyplot as plt
```

n = 10

```python
list_10 = []
n = 10

for i in range(n+1):
  list_10.append(2**i)

print (list_10)
```

```python
plt.plot(list_10)
plt.show()
```
